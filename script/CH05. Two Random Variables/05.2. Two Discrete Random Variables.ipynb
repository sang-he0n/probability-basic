> ## **결합확률질량함수(Joint Probability Mass Function; Joint PMF)**

#### = a function that maps the probability of $ \, 2 \, $ or more discrete random variables occurring simultaneously
#### $ \Rightarrow{} P_{XY}(x,\, y) = P(X=x,\, Y=y) = P(X\cap{}Y) $

#### **(4) 주변확률질량함수(Marginal PMF)** : 
#### 결합확률질량함수에서 **하나의 확률변수에 대해** 다른 이산확률변수 관계없이 정의하는 함수
#### $ \hspace{0.15cm} $ ① 이산확률변수 $ \, X $ 의 확률질량함수
#### $ \hspace{0.6cm} P_{X}(x) = \displaystyle\sum_{y_{k} \in R_{Y}} P(X=x,\, Y=y_{k}) $ 
#### $ \hspace{0.15cm} $ ② 이산확률변수 $ \, Y $ 의 확률질량함수
#### $ \hspace{0.6cm} P_{Y}(y) = \displaystyle\sum_{x_{i} \in R_{X}} P(X=x_{i},\, Y=y) $ 

#### **(WHY?)** (Theorem 5-1) 결합확률변수의 전체 확률의 법칙(Law of Total Probability)
#### 이산확률변수 $ \, X $ 와 분할된(partitioned) 이산확률변수 $ \, Y $ 의 결합합확률변수의 전체확률법칙은 아래와 같음
#### $ \Rightarrow{} P(X=x) = \displaystyle\sum_{y_{k}\in{}R_{Y}}P(X=x,\, Y=y) = \displaystyle\sum_{y_{k}\in{}R_{Y}}P(Y=y_{k})P(X=x|Y=y_{k}) $

> ## **조건부 확률질량함수(Conditional PMF)**

#### **(1) 정의** : 결합확률질량함수에서 **하나의 확률변수가 특정 값을 취했을 때** 다른 확률변수에 대한 함수
#### = a function of another variable given a specific value of one random variable in a joint PMF
#### $ \Rightarrow{} P_{X|Y}(x|y_{k}) = P(X=x|Y=y_{k}) = \frac{P(X=x,\; Y=y_{k})}{P(Y=y_{k})} = \frac{P(\{X=x\} \cap{} \{Y=y_{k}\})}{P(Y=y_{k})}, \;\; $ for any $ x \in{} X(S),\; y_{k} \in{} Y(S) $

#### **(2) 예시** :
#### $ \hspace{0.15cm} $ 결합확률질량함수 $ \; P(X,\; Y) $가 아래의 표로 정의될 때
| $ P(X,\; Y) $ | $ Y = 1 $ | $ Y = 2 $ | $ Y = 5 $ |
|-|-|-|-|
| $ X = 0 $  | $ 0.1 $ | $ 0 $ | $ 0.2 $ |
| $ X = 1 $  | $ 0.5 $ | $ 0.2 $ | $ 0 $ |
#### $ \hspace{0.15cm} $ ① 확률변수 $ \; X = 0 $ 인 확률변수 $ \; Y $ 의 조건부 확률질량함수는?
#### $ \hspace{0.6cm} P(X=x) = \begin{cases} 0.3, \;\; x=0 \\ 0.7, \;\; x=1 \end{cases} $
#### $ \hspace{0.6cm} P(Y=y|X=0) = \begin{cases} \frac{P(X=0,\; Y=y)}{P(X=0)} = \frac{0.1}{0.3} = \frac{1}{3}, \;\; y=1 \\ \frac{P(X=0,\; Y=y)}{P(X=0)} = \frac{0}{0.3} = 0, \;\; y=2 \\ \frac{P(X=0,\; Y=y)}{P(X=0)} = \frac{0.2}{0.3} = \frac{2}{3}, \;\; y=5 \end{cases} $
#### $ \hspace{0.15cm} $ ② 확률변수 $ \; X = 1 $ 인 확률변수 $ \; Y $ 의 조건부 확률질량함수는?
#### $ \hspace{0.6cm} P(Y=y|X=1) = \begin{cases} \frac{P(X=0,\; Y=y)}{P(X=1)} = \frac{0.5}{0.7} = \frac{5}{7}, \;\; y=1 \\ \frac{P(X=0,\; Y=y)}{P(X=1)} = \frac{2}{0.7} = \frac{7}{2}, \;\; y=2 \\ \frac{P(X=0,\; Y=y)}{P(X=1)} = \frac{0}{0.7} = 0, \;\; y=5 \end{cases} $

> ## **조건부 기대값(Conditional Expectation)**

#### **(1) 정의** : 하나의 이산확률변수가 특정 값을 취할 때 다른 이산확률변수의 기대값
#### = a expected value of a discrete random variable given another
#### $ \Rightarrow{} E(X=x|Y=y_{k}) = \displaystyle\sum_{x_{i}\in{}X(S)} x_{i} P(X=x|Y=y_k) $ 

#### **(2) 예시** :
#### $ \hspace{0.15cm} $ ① 공평한 주사위를 $ \; 1 $ 회 던져 주사위눈이 $ \; 3 $ 이하인 사건을 $ \; A $ 라고 정의할 때 $ \; E(X|A) $ 는?
#### $ \hspace{0.6cm} X $ : 주사위눈이 나온 결과
#### $ \hspace{0.6cm} P(X=x|A) = \frac{P(X=x,\; A)}{P(A)} = \begin{cases} \frac{P(X=1,\; A)}{P(A)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}, \;\; x=1 \\ \frac{P(X=2,\; A)}{P(A)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}, \;\; x=2 \\ \frac{P(X=3,\; A)}{P(A)} = \frac{\frac{1}{6}}{\frac{1}{2}} = \frac{1}{3}, \;\; x=3 \end{cases} $ 
#### $ \hspace{0.6cm} \therefore{} \; E(X|A) = xP(X=x|A) = 1 \times{} \frac{1}{3} + 2 \times{} \frac{1}{3} + 3 \times{} \frac{1}{3} = 2 $
#### $ \hspace{0.15cm} $ ② 동전의 앞면이 나올 확률이 $ \; p $ 이며 확률변수 $ \; X $ 는 앞면이 나올 때까지 던진 횟수이고, 첫시행 결과에 대한 조건부확률은?
#### $ \hspace{0.6cm} X $ : 첫 앞면을 볼 때까지의 동전을 던진 횟수
#### $ \hspace{0.6cm} X \sim{} \text{Geometric}(p) $
#### $ \hspace{0.6cm} A = \{ H,\; T \}$ : 첫 번째 시도에서 나온 사건
#### $ \hspace{0.6cm} E(X) = E(X|A=\{H\}) \cdot{} P(H) + E(X|A=\{T\}) \cdot{} P(T) $
#### $ \hspace{1.575cm} = 1 \cdot{} p + (1 + E(X))(1-p) \;\; $ ($ \because{} \; $ each coin toss is independent)
#### $ \hspace{1.575cm} = \frac{1}{p} $

> ## (Theorem 5-2) **전체 기댓값의 법칙(Law of Total Expectation)**

#### 확률변수 $ \; X $ 와 분할된(partitioned) 확률변수 $ \; Y $ 의 결합확률변수 기댓값의 전체확률법칙은 아래와 같다.
#### $ E(X) = \displaystyle\sum_{x_{i}\in{}X(S)} x_{i}P(X=x_{i}) $
#### $ \hspace{0.975cm} = \displaystyle\sum_{x_{i}\in{}X(S)}P(X=x_{i}) \displaystyle\sum_{y_{k}\in{}Y(S)}P(Y=y_{k})P(X=x_{i}|Y=y_{k}) \;\; \cdots{} $
#### $ \hspace{0.975cm} = \displaystyle\sum_{y_{k}\in{}Y(S)} P(Y=y_{k}) \displaystyle\sum_{x_{i}\in{}X(S)} P(X=X_{i})P(X=x_{i}|Y=y_{k}) \;\; \cdots{} $
#### $ \hspace{0.975cm} = \displaystyle\sum_{y_{k}\in{}Y(S)} P(Y=y_{k}) E(X=x_{i}|Y=y_{k}) $

> ## **조건부 분산(Conditional Variance)**

#### = a measure of how far discrete random variables are from the conditional expected value
#### $ \text{var}(X=x|Y=y_{k}) = E[(X-E(X))^{2}|Y=y_{k}] $ 
#### $ \hspace{3.025cm} = $ **[LATEX]**
#### $ \hspace{3.025cm} = E(X^{2}|Y=y_{k}) - E(X)^{2}  $

> ## **이산결합확률변수의 독립(Independnece of Discrete Joint RV)**

#### **(1) 정의** : 한 이산확률변수가 일어날 확률이 다른 이산확률변수가 일어날 확률에 영향을 미치지 않는 관계
#### = Relationship where one discrete RV's probability doesn't impact another's probability
#### $ \Rightarrow{} P(X=x,\; Y=y) = P(X=x)P(Y=y), \;\; $ for all $ \; x \in{} X(S),\; y \in{} Y(S) $
#### $ \Leftrightarrow{} P(X=x|Y=y) = \frac{P(X=x,\; Y=y)}{P(Y=y)} = \frac{P(X=x)P(Y=y)}{P(Y=y)} = P(X=x), \;\; $ for all $ \; x \in{} X(S),\; y \in{} Y(S) $
#### $ \Leftrightarrow{} P(Y=y|X=x) = \frac{P(Y=y,\; X=x)}{P(X=x)} = \frac{P(Y=y)P(X=x)}{P(X=x)} = P(Y=y), \;\; $ for all $ \; x \in{} X(S),\; y \in{} Y(S) $

#### **(PLUS)** 결합 누적분포함수에서도 동일하게 적용됨
#### $ F(X=x,\; Y=y) = F(X=x)F(Y=y), \;\; $ For all $ \; x \in{} X(S),\; y \in{} Y(S) $

#### **(2) 해석** : 
#### 두 확률변수가 독립 관계일 경우, 확률변수 $ \; Y $ 는 확률변수 $ \; X $ 에 대하여 어떠한 정보도 제공하지 않음
#### = 확률변수 $ X $ 의 분포에서 확률변수 $ \; Y $ 는 어떠한 영향을 미치지 않음 

#### **(3) 예시** :
#### 결합확률질량함수 $ \; P(X,\; Y) $가 아래의 표로 정의될 때
| $ P(X,\; Y) $ | $ Y = 1 $ | $ Y = 2 $ | $ Y = 5 $ |
|-|-|-|-|
| $ X = 0 $  | $ 0.1 $ | $ 0 $ | $ 0.2 $ |
| $ X = 1 $  | $ 0.5 $ | $ 0.2 $ | $ 0 $ |
#### $ \hspace{0.15cm} $ ① 확률변수 $ \; X,\; Y $ 는 독립관계인가?
#### $ \hspace{0.6cm} \rightarrow{} P(Y=y|X=x) = P(Y=y) ?\;\; $ for all $ \; x \in{} X(S),\; y \in{} Y(S) $
#### $ \hspace{0.6cm} P(Y=y) = \begin{cases} 0.6, \;\; y=1 \\ 0.2, \;\; y=2 \\ 0.2, \;\; y=5 \end{cases} $
#### $ \hspace{0.6cm} P(Y=y|X=0) = \begin{cases} \frac{1}{3}, \;\; y=1 \\ 0, \;\; y=2 \\ \frac{2}{3}, \;\; y=5 \end{cases}, \;\;\;\; P(Y=y|X=1) = \begin{cases} \frac{5}{7}, \;\; y=1 \\ \frac{2}{7}, \;\; y=2 \\ 0, \;\; y=5 \end{cases} $
#### $ \hspace{0.6cm} \therefore{} \; P(Y=y|X=x) \neq{} P(Y=y) \; $ (not independent)

> ## **이산결합확률변수의 합성함수(Composite Functions of Discrete Joint RV)**

#### **(1) 정의** : 이산결합확률변수를 다른 실수 공간으로 대응시킨 **확률 변수**
#### = a discrete joint random variable mapped to another real space
#### $ \Rightarrow{} Z = g \circ{} (X,Y) = g(X, Y) : S \rightarrow{} \mathbb{R} \;\; $ (※ $ g : \mathbb{R}^2 \rightarrow{} \mathbb{R} $)
<img src="../../png/05.2. Two Discrete Random Variables (10).png" width="50%" height="50%"></img>

#### **(2) 합성함수의 확률** :
#### $ P(Z=z_{m}) = P(g(X, Y)=z_{m}) $
#### $ \hspace{1.825cm} = \displaystyle\sum_{(x_{i},\;y_{k}) \in{} A_{z_{m}} } P_{XY}(x_{i},\; y_{k}) \;\; $ where $ \; A_{z_{m}} = \{ (x_{i},\; y_{k}) \; | \; Z(X,\; Y) = z_{m} \;\; \& \;\; (x_{i},\; y_{k}) \in{} (X,\;Y)(S) \} $

#### **(3) 합성함수의 기댓값** : 
#### $ E(Z) = \displaystyle\sum_{z_{m}\in{}Z(X,\;Y)} z_{m} \displaystyle\sum_{(x_{i},\; y_{k})\in{}A_{z_{k}}} P(X=x_{i},\; Y=y_{k}) $
#### $ \hspace{1cm} = \displaystyle\sum_{z_{m}\in{}Z(X,\;Y)} \displaystyle\sum_{(x_{i},\; y_{k})\in{}A_{z_{k}}} z_{m} P(X=x_{i},\; Y=y_{k}) $
#### $ \hspace{1cm} = \displaystyle\sum_{(x_{i},\; y_{k})\in{}(X,\;Y)(S)} g(X=x_{i},\; Y=y_{k})P(X=x_{i},\; Y=y_{i}) $

#### **(PLUS) 기대값의 선형성 증명** :
#### 두 확률변수 $ \; X,\; Y $ 에 대하여 $ \; g(X,\; Y) = X + Y $ 를 만족할 때
#### $ E(g(X,\; Y)) = \displaystyle\sum_{(x_{i},\; y_{k})\in{}(X,Y)(S)} g(x_{i},\; y_{k}) P(X=x_{i},\; Y=y_{k}) $
#### $ \hspace{2cm} = \displaystyle\sum_{(x_{i},\; y_{k})\in{}X(S) \times{} Y(S)} g(x_{i},\; y_{k}) P(X=x_{i},\; Y=y_{k}) $
#### $ \hspace{2cm} = \displaystyle\sum_{(x_{i},\; y_{k})\in{}X(S) \times{} Y(S)} (x_{i}+y_{k}) P(X=x_{i},\; Y=y_{k}) $
#### $ \hspace{2cm} = \displaystyle\sum_{(x_{i},\; y_{k})\in{}X(S) \times{} Y(S)} x_{i} P(X=x_{i},\; Y=y_{k}) + \displaystyle\sum_{(x_{i},\; y_{k})\in{}X(S) \times{} Y(S)} y_{k} P(X=x_{i},\; Y=y_{k}) $
