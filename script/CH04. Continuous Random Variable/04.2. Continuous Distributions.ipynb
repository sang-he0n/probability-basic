> ## **연속균등 분포(Uniform Distribution)**

#### **(1) 정의** : 특정한 구간 $ \, [a,\, b] $ 에서 가능한 모든 결과가 일정한 확률을 가지는 분포 
#### = a distribution where all outcomes are equally likely within a specific interval $ \, [a,\, b] $
#### $ \Rightarrow{} f(X=x) = \begin{cases} \frac{1}{b-a}, \;\; a \leq{} x \leq{} b \\ 0, \;\;\;\;\;\; x < a \;\; \text{or} \;\; x > b \end{cases} $
<img src="../../png/04.2. Continuous Distribution (1).png" width="35%" height="35%"></img>

#### **(PLUS)** 
#### 확률변수 $ \, X $ 는 **특정 범위 내 모든 값이 균등하게 나타날 때 그 범위**로 정의되며, 그 구간을 모수로 하는 연속균등 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Uniform}(a,\, b) $

#### **(2) 누적분포함수** :
#### $ \Rightarrow{} F(X=x) = P(X\leq{}x) = \begin{cases} 0, \;\;\;\;\;\; x < a \;\; \text{or} \;\; x > b \\ \frac{x-a}{b-a}, \;\; a \leq{} x \leq{} b \\ 1, \;\;\;\;\;\; x > b \end{cases} $ 

#### **(3) 기댓값** : 
#### $ E(X) = \displaystyle\int^{b}_{a} x f(X=x) \text{dx} = \displaystyle\int^{b}_{a} x (\frac{1}{b-a}) \text{dx} $
#### $ \hspace{0.95cm} = \frac{1}{b-a} \begin{bmatrix} \frac{1}{2}x^{2} \end{bmatrix}^{b}_{a} = \frac{b^{2}-a^{2}}{2(b-a)} = \frac{b+a}{2} $

#### **(4) 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = \frac{a^{2}+ab+b^{2}}{3(b-a)} - \frac{b+a}{2} = \frac{(b-a)^{2}}{12}$ 
#### $ \because{} \, ㉠ = \displaystyle\int^{b}_{a} x^{2} (\frac{1}{b-a}) \text{dx} = \frac{1}{b-a} \begin{bmatrix} \frac{1}{3}x^{3} \end{bmatrix}^{b}_{a} = \frac{b^{3}-a^{3}}{3(b-a)} $
#### $ \hspace{0.8cm} = \frac{(b-a)(a^{2}+ab+b^{2})}{3(b-a)} $

> ## **지수 분포(Exponential Distribution)**

#### **(1) 정의** : 어떠한 구간(시간/공간)에서 사건이 발생하는 빈도를 정의하는 확률변수가 따르는 분포
#### = a distribution that defines the frequency of events occurring within a specific interval(time/space)
#### $ \Rightarrow{} f(X=x) = \begin{cases} \lambda{}e^{-\lambda{}x} = \frac{\lambda{}}{e^{\lambda{}x}}, \;\; x \geq{} 0 \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x < 0 \end{cases} \;\; $ with $ \; \lambda{} > 0 $
<img src="../../png/04.2. Continuous Distribution (2).png" width="40%" height="40%"></img>

#### **(PLUS)** 
#### 확률변수 $ \, X $ 는 **어떠한 사건이 발생할 빈도 혹은 간격**으로 정의되며, 사건의 단위 구간동안의 발생률 $ \, \lambda{} $ 를 모수로 하는 지수 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Exponential}(\lambda{}) $

#### **(2) 누적분포함수** :
#### $ \Rightarrow{} F(X=x) = P(X\leq{}x) = \begin{cases} \displaystyle\int^{x}_{0} \lambda{}e^{-\lambda{}u} \text{du} = 1 - e^{-\lambda{}x}, \;\; x \geq{} 0 \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x < 0 \end{cases} $ 

#### **(PLUS) 지수함수 $ e^{g(x)} $ 의 미분과 적분**
#### $ \hspace{0.15cm} \cdot{} \, $ 미분 : $ \frac{\text{d}}{\text{dx}}e^{g(x)} = g'(x) e^{g(x)} \;\; $ ($ \because{} \; \frac{\text{d}}{\text{dx}}e^{x} = e^{x} $)
#### $ \hspace{0.15cm} \cdot{} \, $ 적분 : $ \displaystyle\int{}e^{g(x)}\text{dx} = \frac{1}{g'(x)} e^{g(x)} + C \;\; $ ($ \because{} \; \displaystyle\int{}e^{x}\text{dx} = e^{x} + C $)  

#### **(3) 기댓값** : 
#### $ E(X) = \displaystyle\int^{\infty{}}_{0} x f(X=x) \text{dx} = \displaystyle\int^{\infty{}}_{0} x \lambda{}e^{-\lambda{}x} \text{dx} $
#### $ \hspace{0.95cm} = \begin{bmatrix} x \cdot{} -\frac{\lambda{}}{\lambda{}} e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} - \displaystyle\int^{\infty{}}_{0} (-\frac{\lambda{}}{\lambda{}} e^{-\lambda{}x}) \text{dx} = \begin{bmatrix} - x e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} + \displaystyle\int^{\infty{}}_{0} e^{-\lambda{}x} \text{dx} $
#### $ \hspace{0.95cm} = \begin{bmatrix} - x e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} + \begin{bmatrix} -\frac{1}{\lambda{}} e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} $
#### $ \hspace{0.95cm} = (\displaystyle\lim_{x\rightarrow{}\infty{}}\frac{x}{e^{\lambda{}{}x}} - 0 ) -\frac{1}{\lambda{}}(\displaystyle\lim_{x\rightarrow{}\infty{}}\frac{1}{e^{\lambda{}{}x}} - 1) = (0 - 0) - \frac{1}{\lambda{}}(0 - 1 )$
#### $ \hspace{0.95cm} = \frac{1}{\lambda{}} $

#### **(PLUS) 부분적분법 정의** :
#### $ \displaystyle\int^{b}_{a} f(x)g(x) \text{dx} = \begin{bmatrix} f(x) \displaystyle\int{} g(x) \text{dx} \end{bmatrix}^{b}_{a} - \displaystyle\int^{b}_{a} f'(x) (\displaystyle\int{} g(x) \text{dx}) \text{dx} $

#### **(WHY?)**
#### $ \frac{\text{d}}{\text{dx}} \{ f(x)k(x) \} = f'(x)k(x) + f(x)k'(x) \;\; $ ($ \because{} \; $ 곱의 미분법)
#### $ \displaystyle\int{} (\frac{\text{d}}{\text{dx}} \{ f(x)k(x) \}) \text{dx} = \displaystyle\int{} (f'(x)k(x) + f(x)k'(x)) \text{dx} $
#### $ f(x)k(x) = \displaystyle\int{} f'(x)k(x) \text{dx} + \displaystyle\int{} f(x)k'(x) \text{dx} $
#### $ \displaystyle\int{} f(x)k'(x) \text{dx} = f(x)k(x) - \displaystyle\int{} f'(x)k(x) \text{dx} $
#### $ \displaystyle\int{} f(x)g(x) \text{dx} = f(x) \displaystyle\int{} g(x) \text{dx} - \displaystyle\int{} f'(x) (\displaystyle\int{} g(x) \text{dx}) \text{dx} \;\;\;\; $ where $ \; g(x) = k'(x) = \frac{\text{d}}{\text{dx}} k(x) $

#### **(4) 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = \frac{2}{\lambda{}^{2}} - \frac{1}{\lambda{}^{2}} = \frac{1}{\lambda{}^{2}} $ 
#### $ \because{} \; ㉠ = \displaystyle\int^{\infty{}}_{0} x^{2} \lambda{}e^{-\lambda{}x} \text{dx} = \begin{bmatrix} -x^{2} e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} + \displaystyle\int^{\infty{}}_{0} 2 x e^{-\lambda{}x} \text{dx} $
#### $ \hspace{0.8cm} = \begin{bmatrix} -x^{2} e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} - 2 \{ \begin{bmatrix} -\frac{1}{\lambda{}} x e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} - \displaystyle\int^{\infty{}}_{0} (-\frac{1}{\lambda{}}e^{-\lambda{}x}) \text{dx} \} $
#### $ \hspace{0.8cm} = (0 - 0) - 2 \{ (0 - 0) - \frac{1}{\lambda{}} \begin{bmatrix} -\frac{1}{\lambda{}}e^{-\lambda{}x} \end{bmatrix}^{\infty{}}_{0} \} $
#### $ \hspace{0.8cm} = - 2 \begin{bmatrix} - \frac{1}{\lambda{}} (0 - \frac{1}{\lambda{}}) \end{bmatrix} = \frac{2}{\lambda{}^{2}} $

#### **(5) 성질** :
#### $ \hspace{0.15cm} $ ① 무기억성질(memoryless) : 이전 구간의 확률이 다음 구간의 확률에 영향을 미치지 않음
#### $ \hspace{0.6cm} P(X > s+t | X > t ) = P(X > s) $

> ## **감마 분포(Gamma Distribution)**

#### **(1) 정의** : 어떠한 구간(시간/공간)에서 사건이 발생하는 **누적** 빈도를 정의하는 확률변수가 따르는 분포
#### = a distribution for the cumulative frequency of events over time or space
#### $ \Rightarrow{} f(X=x) = \begin{cases} \frac{\lambda{}^{\alpha{}}x^{\alpha{}-1}e^{-\lambda{}x}}{\Gamma{}(\alpha{})}, \;\; x > 0 \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\; x \leq{} 0 \end{cases} $
<img src="../../png/04.2. Continuous Distribution (3).png" width="40%" height="40%"></img>

#### **(PLUS) 감마 함수(Gamma Function)의 정의** : 팩토리얼을 실수 범위까지 확장한 함수
#### $ \hspace{0.15cm} \cdot{} \, $ 자연수 범위 : $ \, \Gamma{}(n) = (n-1)!, \;\; $ for all $ \, n \in{} \mathbb{N} $
#### $ \hspace{0.15cm} \cdot{} \, $ 비음수 실수 범위 : $ \, \Gamma{}(\alpha{}) = \displaystyle\int^{\infty{}}_{0} x^{\alpha{}-1} e^{-x} \text{dx} , \;\; $ for all $ \, \alpha{} > 0 $

#### **(WHY?)**
#### $ \Gamma{}(\alpha{}+1) = \displaystyle\int^{\infty{}}_{0} x^{\alpha{}} e^{-x} \text{dx} = \begin{bmatrix} -x^{\alpha{}} e^{-x} \end{bmatrix}^{\infty{}}_{0} + \displaystyle\int^{\infty{}}_{0} \alpha{} \cdot{} x^{\alpha{}-1} (-1) \cdot{} e^{-x} \text{dx} $ 
#### $ \hspace{1.425cm} = (0 - 0) + \alpha{} \displaystyle\int^{\infty{}}_{0} x^{\alpha{}-1} e^{-x} \text{dx} \;\; $ ($ \because{} \, \displaystyle\lim_{x\rightarrow{}\infty{}} -\frac{x^{\alpha{}}}{e^{x}} = 0 $ ; 로피탈의 정리)
#### $ \hspace{1.425cm} = \alpha{} \Gamma{}(\alpha{}) $
#### $ \therefore{} \, \Gamma{}(n) = (n-1) \Gamma{}(n-1) = (n-1) (n-2) \Gamma{}(n-2) = \cdots{} = (n-1)! $

#### **(PLUS)** 
#### 확률변수 $ \, X $ 는 **[CONTENTS]** 으로 정의되며, **[CONTENTS]** 을 모수로 하는 감마 분포를 따름 
#### $ \Rightarrow{} X \sim{} \text{Gamma}(\alpha{},\, \lambda{}) $

#### **(2) 누적분포함수** :
#### $ \Rightarrow{} F(X=x) = P(X\leq{}x) = \begin{cases} \displaystyle\int^{x}_{0} \frac{\lambda{}^{\alpha{}}x^{\alpha{}-1}e^{-\lambda{}x}}{\Gamma{}(\alpha{})}, \;\; x \geq{} 0 \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x < 0 \end{cases} $ 

#### **(3) 기댓값** : 
#### $ E(X) = \displaystyle\int^{\infty{}}_{0} x f(X=x) \text{dx} = \displaystyle\int^{\infty{}}_{0} x \frac{\lambda{}^{\alpha{}}x^{\alpha{}-1}e^{-\lambda{}x}}{\Gamma{}(\alpha{})} \text{dx} = \frac{\lambda{}}{\Gamma{}(\alpha{})} \displaystyle\int^{\infty{}}_{0} x^{\alpha{}}e^{-\lambda{}x} \text{dx} $
#### $ \hspace{0.95cm} = \frac{\lambda{}}{\Gamma{}(\alpha{})} \frac{\Gamma{}(\alpha{}+1)}{\lambda{}^{\alpha{}+1}} = \frac{\lambda{}^{\alpha{}}}{\Gamma{}(\alpha{})} \frac{\alpha{}\Gamma{}(\alpha{})}{\lambda{}^{\alpha{}+1}} = \frac{\alpha{}}{\lambda{}} $ 

#### **(WHY?)**
#### $ \Gamma{}(\alpha{}) = \displaystyle\int^{\infty{}}_{0} x^{\alpha{}-1} e^{-x} \text{dx}, \;\; $ for $ \; \alpha{} > 0 $
#### $ \hspace{0.825cm} = \displaystyle\int^{\infty{}}_{0} (\lambda{}u)^{\alpha{}-1} e^{-\lambda{}u} \lambda{}\text{du}, \;\; $ where $ \; x = \lambda{} u,\; \frac{\text{dx}}{\text{du}} = \lambda{},\; \text{dx} = \lambda{}\text{du} $
#### $ \hspace{0.825cm} = \lambda{}^{\alpha{}} \displaystyle\int^{\infty{}}_{0} u^{\alpha{}-1}e^{-\lambda{}u} \text{du} $
#### $ \therefore{} \; \frac{\Gamma{}(\alpha{})}{\lambda{}^{\alpha{}}} = \displaystyle\int^{\infty{}}_{0} u^{\alpha{}-1}e^{-\lambda{}u} \text{du} $

#### **(4) 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = \frac{(\alpha{}+1)\alpha{}}{\lambda{}^{2}} - (\frac{\alpha{}}{\lambda{}})^{2} = \frac{\alpha{}}{\lambda{}^{2}} $
#### $ \because{} \; ㉠ = \displaystyle\int^{\infty{}}_{0} x^{2} \frac{\lambda{}^{\alpha{}}x^{\alpha{}-1}e^{-\lambda{}x}}{\Gamma{}(\alpha{})} \text{dx} = \frac{\lambda{}}{\Gamma{}(\alpha{})} \displaystyle\int^{\infty{}}_{0} x^{\alpha{}+1}e^{-\lambda{}x} \text{dx} $ 
#### $ \hspace{0.8cm} = \frac{\lambda{}}{\Gamma{}(\alpha{})} \frac{\Gamma{}(\alpha{}+2)}{\lambda{}^{\alpha{}+2}} = \frac{(\alpha{}+1)\alpha{}}{\lambda{}^{2}} $

> ## **표준 정규분포(Standard Normal Distribution)**

#### **(1) 정의** : 평균이 $ \, 0 $ 이고 분산이 $ \, 1 $ 이며 **대칭적인 종모양을 갖는** 확률변수가 따르는 분포
#### = a distribution followed by a random variable with mean of $ \, 0 \, $ and deviation of $ \, 1 $
#### $ \Rightarrow{} f(Z=z) = \frac{1}{\sqrt{2\pi{}}}e^{-\frac{z^{2}}{2}}, \;\; -\infty{} \leq{} z \leq{} \infty{} $
<img src="../../png/04.2. Continuous Distribution (4).png" width="40%" height="40%"></img>

#### **(WHY?)**
#### **[LATEX]** (가우스함수의 일반화)

#### **(PLUS)** 
#### 확률변수 $ \, Z $ 는 **평균이 $ \, 0 $ , 분산이 $ \, 1 $ 인 종모양의 곡선**으로 정의되며, $ \, Z $ 의 평균과 분산을 모수로 하는 표준 정규분포를 따름
#### $ \Rightarrow{} Z \sim{} \text{N}(\text{E}(Z) = 0,\, \text{var}(Z) = 1) $

#### **(2) 누적분포함수** :
#### $ \Rightarrow{} \varPhi{}(Z=z) = F(Z=z) = P(z\leq{}z) = \displaystyle\int^{z}_{-\infty{}} \frac{1}{\sqrt{2\pi{}}}e^{-\frac{u^{2}}{2}} \text{du}, \;\; -\infty{} \leq{} z \leq{} \infty{} $ 

#### **(PLUS)** 표준정규분포의 누적분포함수는 초등함수로 표현이 안되지만, 자주 쓰이므로 **표준정규분포표**를 통해 값을 근사함

#### **(PLUS) 초등함수(Elementary Function)의 정의** :
#### **[CONTENTS]**

#### **(3) 기댓값** : 
#### $ E(Z) = \displaystyle\int^{\infty{}}_{-\infty{}} z f(Z=z) \text{dz} = \displaystyle\int^{\infty{}}_{-\infty{}} z \frac{1}{\sqrt{2\pi{}}}e^{-\frac{z^{2}}{2}} \text{dz} $
#### $ \hspace{0.95cm} = \frac{1}{\sqrt{2\pi{}}}(-\displaystyle\int^{\infty{}}_{0} z e^{-\frac{z^{2}}{2}} \text{dz} + \displaystyle\int^{\infty{}}_{0} z e^{-\frac{z^{2}}{2}} \text{dz}) \;\; $ ($ \because{} \, z $ 가 곱해졌고 자연상수 지수가  $ \, z^{2} $ 이므로 기함수)
#### $ \hspace{0.95cm} = \frac{1}{\sqrt{2\pi{}}}(-\displaystyle\int^{\infty{}}_{0} e^{-u} \text{du} + \displaystyle\int^{\infty{}}_{0} e^{-u} \text{du}), \;\; $ where $ \, u = \frac{z^{2}}{2},\; \frac{\text{du}}{\text{dz}} = z, \;\text{du} = z\text{dz} $  
#### $ \hspace{0.95cm} = \frac{1}{\sqrt{2\pi{}}}(\begin{bmatrix} e^{-u} \end{bmatrix}^{\infty{}}_{0} - \begin{bmatrix} e^{-u} \end{bmatrix}^{\infty{}}_{0}) $
#### $ \hspace{0.95cm} = \frac{1}{\sqrt{2\pi{}}}(0 - 0) = 0 $

#### **(4) 분산** :
#### $ \text{var}(Z) = \underbrace{E(Z^{2})}_{\text{㉠}} - E(Z)^{2} = 1 - 0 = 1 $ 
#### $ \because{} \; ㉠ = \displaystyle\int^{\infty{}}_{-\infty{}} z^{2} \frac{1}{\sqrt{2\pi{}}}e^{-\frac{z^{2}}{2}} \text{dz} $ 
#### $ \hspace{0.8cm} = 2 \displaystyle\int^{\infty{}}_{0} z^{2} \frac{1}{\sqrt{2\pi{}}}e^{-\frac{z^{2}}{2}} \text{dz} \;\; $ ($ \because{} \; z^{2} $ 이 곱해졌고 자연상수 지수가  $ \; z^{2} $ 이므로 우함수)
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{2\pi{}}} \displaystyle\int^{\infty{}}_{0} z \cdot{} e^{-\frac{z^{2}}{2}} \cdot{} z \text{dz} $
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{2\pi{}}} \displaystyle\int^{\infty{}}_{0} \sqrt{2u} e^{-u} \text{du} \;\; $ where $ \; u = \frac{z^{2}}{2},\; \text{du} = z\text{dz} $ 
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{\pi{}}} \displaystyle\int^{\infty{}}_{0} u^{\frac{1}{2}} e^{-u} \text{du} $
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{\pi{}}} \Gamma{}(\frac{3}{2}) \;\; $ ($ \because{} \; \Gamma{}(\alpha{}) = \displaystyle\int^{\infty{}}_{0} x^{\alpha{}-1} e^{-x} \text{dx} $) 
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{\pi{}}} \cdot{} \frac{1}{2} \cdot{} \Gamma{}(\frac{1}{2}) \;\; $ ($ \because{} \Gamma{}(\alpha{}+1) = \alpha{}\Gamma{}(\alpha{})\; $)
#### $ \hspace{0.8cm} = \frac{2}{\sqrt{\pi{}}} \cdot{} \frac{1}{2} \cdot{} \sqrt{\pi{}} = 1 $

#### **(PLUS) 우함수와 기함수의 정의** : 
#### $ \hspace{0.15cm} \cdot{} \, $ 우함수(even function) : 수직축에 대하여 대칭 $ \, \rightarrow{} f(x) = f(-x) $
#### $ \hspace{0.15cm} \cdot{} \, $ 기함수(odd function) : 원점에 대하여 대칭 $ \, \rightarrow{} f(x) = -f(-x) $

> ## **정규분포(Normal Distribution; Gaussian Distribution)**

#### **(1) 정의** : 평균이 $ \, \mu{} $ 이고 분산이 $ \, \sigma{}^{2} $ 이며 **대칭적인 종모양을 갖는** 확률변수가 따르는 분포
#### = a distribution followed by a random variable with mean of $ \, \mu{} \, $ and deviation of $ \, \sigma{}^{2} $
#### = 표준정규분포를 따르는 확률 변수 $ \, Z $ 에 대하여 확률변수 $ \, X  = \sigma{} X + \mu{} $ 가 따르는 분포
#### $ \Rightarrow{} f(X=x) = \frac{1}{\sqrt{2\cdot{}\pi{}\cdot{}\sigma{}^{2}}}e^{-\frac{(x-\mu{})^{2}}{2\cdot{}\sigma{}^{2}}}, \;\; -\infty{} \leq{} x \leq{} \infty{} $

#### **(PLUS)** 
#### 확률변수 $ \, X $ 는 **평균이 $ \, \mu{} $ , 분산이 $ \, \sigma{}^{2} $ 인 종모양의 곡선**으로 정의되며, $ \, X $ 의 평균과 분산을 모수로 하는 정규분포를 따름
#### $ \Rightarrow{} X \sim{} \text{N}(\text{E}(X) = \mu{},\, \text{var}(X) = \sigma{}^{2}) $

#### **(2) 누적분포함수** :

#### **(3) 기댓값** : 
#### $ E(X) = E(\sigma{}Z+\mu{}) = E(\sigma{}Z) + \mu{} = \sigma{}E(Z) + \mu{} = \mu{} $

#### **(4) 분산** :
#### $ \text{var}(X) = \text{var}(\sigma{}Z+\mu{}) = \sigma{}^{2} \text{var}(Z) = \sigma{}^{2} $
