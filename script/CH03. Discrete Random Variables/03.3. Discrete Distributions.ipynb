> ## **베르누이 분포(Bernoulli Distribution)**

#### **(1) 정의** : $ \; 1 $ 회 시행하는 실험이 $ \; 2 $ 가지 결과(성공/실패)만 발생할 때 특정 결과가 나올 확률이 $ \; p $ 인 확률변수가 따르는 분포
#### = a distribution that results from a single trial with a probability $ \; p \; $ of one specific outcome among two possible outcomes
#### $ \Rightarrow{} P(X=x) = \begin{cases} p, \;\;\;\;\;\;\; x = 1 \\ 1-p,\; x = 0 \\ 0, \;\;\;\;\;\;\; x \not\in{} \{0,\; 1\} \end{cases} $
<img src="../../png/03.3. Discrete Distributions (1).png" width="35%" height="35%"></img>

#### **(PLUS)** 이 때 확률 변수 $ \; X $ 는 **$ \; 2 $개의 결과 중 특정 결과의 여부**로 정의되며,
#### 특정 결과의 확률 $ \; p $ 를 모수(parameter)로 하는 베르누이 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Bernoulli}(p) $

#### **(2) 분포의 평균** : 
#### $ E(X) = \displaystyle\sum^{1}_{x=0} P(X=x) = 0 \times{} P(X=0) + 1 \times{} P(X=1) = p $

#### **(3) 분포의 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = p - p^{2} = p(1-p) $ 
#### $ \because{} \; ㉠ = (0^{2}\times{}P(X=0)+1^{2}\times{}P(X=1)) = p $

#### **(4) 분포의 성질**
#### $ \hspace{0.15cm} $ ① 결과의 이진성
#### $ \hspace{0.15cm} $ ② 단일 실험

#### **(5) 예시** : 
#### $ \hspace{0.15cm} $ ① 불공평한 동전 $ \; 1 $ 회 던졌을 때, 확률변수 $ \; X $ 는 앞면이 나온 경우를 $ \; 1 $ 이라고 정의하면 
#### $ \hspace{0.55cm} $ (단, 앞면이 확률을 $ \; 30\% $ 라고 정의함)
#### $ \hspace{0.6cm} S = \{ \text{H},\; \text{T} \} $
#### $ \hspace{0.6cm} X = \{ 1,\; 0 \} \sim{} \text{Bernoulli}(p=0.3) $
#### $ \hspace{0.6cm} P(X=x) = \begin{cases} 0.3, \;\;\;\;\;\;\; x = 1 = X(\{ H \}) \\ 0.7, \;\;\;\;\;\;\; x = 0 = X(\{ T \}) \\ 0, \;\;\;\;\;\;\;\;\;\; x \not\in{} \{0,\; 1\} = X(\{ H,\; T \}) \end{cases} $
#### $ \hspace{0.15cm} $ ② 불공평한 주사위 $ \; 1 $ 회 던졌을 때, 확률변수 $ \; Y $ 는 $ \; 3 $ 이하의 눈이 나온 경우를 $ \; 1 $이라고 정의하면
#### $ \hspace{0.55cm} $ (단, $ \; 3 $ 이하의 눈이 나올 확률을 $ \; 60\% $ 라고 정의함)
#### $ \hspace{0.6cm} S = \{ \text{1},\; \text{2},\; \text{3},\; \text{4},\; \text{5},\; \text{6} \} $
#### $ \hspace{0.6cm} Y = \{ 1,\; 1,\; 1,\; 0,\; 0,\; 0 \} = \{ 1,\; 0 \} \sim{} \text{Bernoulli}(p=0.6) $
#### $ \hspace{0.6cm} P(Y=y) = \begin{cases} 0.6, \;\;\;\;\;\;\; y = 1 = Y(\{ 1,\; 2,\; 3 \}) \\ 0.4, \;\;\;\;\;\;\; y = 0 = Y(\{ 4,\; 5,\; 6 \}) \\ 0, \;\;\;\;\;\;\;\;\;\; y \not\in{} \{ 0,\; 1 \} = Y(\{ 1,\; 2,\; 3,\; 4,\; 5,\; 6\}) \end{cases} $

> ## **이항 분포(Binomial Distribution)**

#### **(1) 정의** : $ \; n $ 회 시행하는 실험이 $ \; 2 $ 가지 결과만 발생할 때 특정 결과가 나올 확률이 $ \; p $ 인 확률변수가 따르는 분포
#### = a distribution that results from several trial with a probability $ \; p \; $ of one specific outcome among two possible outcomes
#### $ \Rightarrow{} P(X=x) = \begin{cases} C^{n}_{x} p^{x} (1-p)^{n-x}, \;\; x \in{} \{0,\; 1,\; \cdots{},\; n \} \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x \not\in{} \{0,\; 1,\; \cdots{},\; n \} \end{cases} $
<img src="../../png/03.3. Discrete Distributions (2).png" width="40%" height="40%"></img>

#### **(PLUS)** 이 때 확률 변수 $ \; X $ 는 **$ \; n $ 회 시행 중 특정 결과가 관측된 횟수**로 정의되며, 
#### 시행 횟수 $ \; n $ 과 특정 결과의 확률 $ \; p $ 를 모수로 하는 이항 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Binomial}(n,\; p) $

#### **(2) 분포의 평균** : 
#### $ E(X) = \displaystyle\sum^{n}_{x=0} xP(X=x) = \displaystyle\sum^{n}_{x=0} xC^{n}_{x} p^{x} q^{n-x} \;\; $ where $ \; q = 1 - p $
#### $ \hspace{0.95cm} = 0 + \displaystyle\sum^{n}_{x=1} xC^{n}_{x} p^{x} q^{n-x} $ 
#### $ \hspace{0.95cm} = \displaystyle\sum^{n}_{x=1} x {{n!}\over{(n-x)!x!}} p^{x} q^{n-x} = \displaystyle\sum^{n}_{x=1} {{n!}\over{(n-x)!(x-1)!}} p^{x} q^{n-x} $
#### $ \hspace{0.95cm} = \displaystyle\sum^{n}_{x=1} {{(n)(n-1)!}\over{(n-x)!(x-1)!}} p^{x} q^{n-x} = n \displaystyle\sum^{n}_{x=1} {{(n-1)!}\over{(n-x)!(x-1)!}} p^{x} q^{n-x} $ 
#### $ \hspace{0.95cm} = n \displaystyle\sum^{n-1}_{y=0} {{(n-1)!}\over{(n-x)!y!}} p^{y+1} q^{n-(y+1)} \;\; $ where $ \; y = x - 1, \;\; Y \sim{} \text{Binomial}(n-1, p) $ 
#### $ \hspace{0.95cm} = n \displaystyle\sum^{n-1}_{y=0} {{(n-1)!}\over{(n-x)!y!}} (p)(p)^{y} q^{n-y-1} = n p \displaystyle\sum^{n-1}_{y=0} {{(n-1)!}\over{(n-x)!y!}} p^{y} q^{(n-1)-y} $
#### $ \hspace{0.95cm} = n p \displaystyle\sum^{n-1}_{y=0} C^{n-1}_{y} p^{y} q^{(n-1)-y} $
#### $ \hspace{0.95cm} = n p (p+q)^{n-1} \;\; $ ($ \because{} \; $ 이항정리)
#### $ \hspace{0.95cm} = n p \;\; $ ($ \because{} \; p + q = 1 $)

#### **(3) 분포의 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = n(n-1) p^{2} + np - (np)^{2} = np(1-p) $ 
#### $ \because{} \; ㉠ = E(X(X-1)) + E(X) $
#### $ \hspace{0.9cm} = \displaystyle\sum^{n}_{x=0} x(x-1)C^{n}_{x} p^{x} q^{n-x} + np \;\; $ where $ \; q = 1 - p $
#### $ \hspace{0.9cm} = \displaystyle\sum^{n}_{x=0} x(x-1)C^{n}_{x} p^{x} q^{n-x} + np = \displaystyle\sum^{n}_{x=0} x(x-1){{n!}\over{(n-x)!x!}} p^{x} q^{n-x} + np $
#### $ \hspace{0.9cm} = \displaystyle\sum^{n}_{x=0} {{n!}\over{(n-x)!(x-2)!}} p^{x} q^{n-x} + np = \displaystyle\sum^{n}_{x=0} n(n-1){{(n-2)!}\over{(n-x)!(x-2)!}} p^{x} q^{n-x} + np $
#### $ \hspace{0.9cm} = n(n-1) \displaystyle\sum^{n}_{x=0} C^{n-2}_{x-2} p^{x} q^{n-x} + np $
#### $ \hspace{0.9cm} = n(n-1) \displaystyle\sum^{n-2}_{y=0} C^{n-2}_{y} p^{y+2} q^{n-y-2} + np \;\; $ where $ \; y = x-2, \;\; Y \sim{} \text{Binomial}(n-2, p)  $
#### $ \hspace{0.9cm} = n(n-1) \displaystyle\sum^{n-2}_{y=0} C^{n-2}_{y} p^{y} \cdot{} p^{2} q^{n-y-2} + np = n(n-1) p^{2} \displaystyle\sum^{n-2}_{y=0} C^{n-2}_{y} p^{y} q^{n-y-2} + np $
#### $ \hspace{0.9cm} = n(n-1) p^{2} + np \;\; $ ($ \because{} \; $ 이항정리)

#### **(PLUS) 이항 분포와 베르누이 분포의 관계** :
#### 확률 변수 $ \; X_{1},\; X_{2},\; \cdots{},\; X_{n} $ 이 특정 결과의 확률 $ \; p $ 를 모수로 삼는 **독립적인** 베르누이 확률변수라면, 
#### 확률 변수 $ X = X_{1},\; X_{2},\; \cdots{},\; X_{n} $ 은 시행 횟수 $ \; n $, 특정 결과의 확률 $ \; p $ 를 모수로 삼는 이항 분포를 따름
#### $ \Rightarrow{}  X_{i} \sim{} \text{Bernoulli}(p) \;\; $ ($ i = 1,\; 2,\; \cdots{},\; n $) $ \; \rightarrow{} \; X_{1} + X_{2} + \cdots{} + X_{n} = X \sim{} \text{Binomial}(n,\; p) $ 
#### $ \hspace{1.2cm} (\text{independent}) \hspace{3.25cm} \; P(X_{1}\cap{}X_{1}\cap{}\cdots{}\cap{}X_{n}) = P(X) \;\; $ ($ \because{} \; $ and 조건)
#### $ \hspace{6.6cm} \; P(X_{1})P(X_{1})\cdots{}P(X_{n}) \; = $
#### 또한 기대값, 분산의 성질에 의해 아래와 같음
#### $ E(X) = E(X_{1}) + E(X_{2}) + \cdots{} E(X_{n}) = np $
#### $ \text{var}(X) = \text{var}(X_{1}) + \text{var}(X_{2}) + \cdots{} + \text{var}(X_{n}) = np(1-p) \;\; $ where $ \; X_{i} \sim{} \text{Bernoulli}(p) $

#### **(PLUS-EXAMPLE)** 
#### 확률 변수 $ X \sim{} \text{Binomial}(n,\; p) $ 이고, $ Y \sim{} \text{Binomial}(m,\; p) $ 일 때 확률 변수 $ Z = X + Y $ 의 확률밀도함수는
#### $ \hspace{0.3cm} X = \displaystyle\sum^{n}_{i=1} X_{i}, \;\; X_{i} \sim{} \text{Bernoulli}(p) $
#### $ \hspace{0.3cm} Y = \displaystyle\sum^{m}_{k=1} Y_{k}, \;\; Y_{k} \sim{} \text{Bernoulli}(p) $
#### $ \hspace{0.3cm} Z = X + Y = \displaystyle\sum^{n}_{i=1} X_{i} + \displaystyle\sum^{m}_{k=1} Y_{k}, \;\; Z \sim{} \text{Binomial}(m+n,\; p) $ 
#### $ \hspace{0.3cm} P(Z=z) = \begin{cases} C^{m+n}_{z} (p)^{z}(1-p)^{m+n-z}, \;\; z \in{} \{ 0,\; 1,\; \cdots{},\; m+n \} \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; z \not\in{} \{ 0,\; 1,\; \cdots{},\; m+n \} \end{cases} $

> ## **기하 분포(Geometric Distribution)**

#### **(1) 정의** : 실험에서 성공확률이 $ \; p $ 인 베르누이 시행에서 특정 결과가 $ \; 1 $ 회 나올 때까지 시행할 때, 그 확률 변수가 따르는 분포
#### = a distribution of trials until the first success in Bernoulli trials with probability $ \; p $
#### $ \Rightarrow{} P(X=x) = \begin{cases} (1-p)^{x-1}p, \;\; x = \{ 1,\; 2,\; \cdots{} \} \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x \not\in{} \{ 1,\; 2,\; \cdots{} \} \end{cases} $
<img src="../../png/03.3. Discrete Distributions (3).png" width="40%" height="40%"></img>

#### **(PLUS)** 이 때 확률 변수 $ \; X $ 는 **특정 결과가 $ \; 1 $ 회 나올 때까지 시행하는 횟수**로 정의되며, 
#### 특정 결과의 확률 $ \; p $ 를 모수로 하는 기하 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Geometric}(p) $

#### **(2) 분포의 평균** : 
#### $ E(X) = \displaystyle\sum^{\infty{}}_{x=0} P(X=x) = \displaystyle\sum^{\infty{}}_{x=1} x p q^{x-1} \;\; $ where $ \; q = 1 - p $
#### $ \hspace{1cm} = p \displaystyle\sum^{\infty{}}_{x=1} x q^{x-1} $
#### $ \hspace{1cm} = p \displaystyle\sum^{\infty{}}_{x=1} {{\text{d}}\over{\text{d}q}} q^{x}  = p {{\text{d}}\over{\text{d}q}} (\displaystyle\sum^{\infty{}}_{x=1}{q^{x}}) = p {{\text{d}}\over{\text{d}q}} ({{1}\over{1-q}}) \;\; $ ($ \because{} \; $ 공비가 $ \; 1 \; $ 이하인 등비수열의 무한 합)
#### $ \hspace{1cm} = p {{1}\over{(1-q)^{2}}} = p {{1}\over{(p)^{2}}} = {{1}\over{p}} $

#### **(3) 분포의 분산** :
#### $ \text{var}(X) = \underbrace{E(X^{2})}_{\text{㉠}} - E(X)^{2} = {{2}\over{p^{2}}} - {{1}\over{p}} - {{1}\over{p^{2}}} = {{1-p}\over{p^{2}}} $ 
#### $ \because{} \; ㉠ = \displaystyle\sum^{\infty{}}_{x=1} x^{2} p q^{x-1} \;\; $ where $ \; q = 1 - p $
#### $ \hspace{0.85cm} = \displaystyle\sum^{\infty{}}_{x=1}x(x+1)pq^{x-1} - \displaystyle\sum^{\infty{}}_{x=1}xpq^{x-1} $
#### $ \hspace{0.85cm} = \displaystyle\sum^{\infty{}}_{x=1} x(x+1)pq^{x-1} - E(X) = p \displaystyle\sum^{\infty{}}_{x=1}x(x+1)q^{x-1} - E(X) $
#### $ \hspace{0.85cm} = p \displaystyle\sum^{\infty{}}_{x=1}{{\text{d}}\over{\text{d}q}}[(x+1)q^{x}] - E(X) = p {{\text{d}}\over{\text{d}q}}[\underbrace{\displaystyle\sum^{\infty{}}_{x=1}(x+1)q^{x}}_{\text{ⓐ}}] - E(X) $
#### $ \hspace{0.85cm} = p {{\text{d}}\over{\text{d}q}}[{{1}\over{(1-q)^{2}}}-1] - E(X) $
#### $ \hspace{0.85cm} = p {{2}\over{(1-q)^{3}}} - E(X) = {{2}\over{p^{2}}} - {{1}\over{p}} $

#### **(PLUS) ⓐ 유도** :
#### $ ⓐ = 2q + 3q^{2} + \cdots{}, \;\; ⓐ\times{}(1-q) = 2q + q^{2} + \cdots{} = q + q + q^{2} + \cdots{} = q + {{q}\over{1-q}}$ 
#### $ ⓐ = {{q}\over{(1-q)}} + {{q}\over{(1-q)^{2}}} = {{1}\over{(1-q)^{2}}} - 1 $

> ## **파스칼 분포(Pascal Distribution; Negative Binomial Distribution)**

#### **(1) 정의** : 실험에서 성공확률이 $ \; p $ 인 베르누이 시행에서 특정 결과가 $ \; n $ 회 나올 때까지 시행할 때, 그 확률 변수가 따르는 분포
#### = of trials until the $ \; n $ -th success in Bernoulli trials with probability
#### $ \Rightarrow{} P(X=x) = \begin{cases} C^{x-1}_{n-1}(1-p)^{x-n}(p)^{n}, \;\; x = \{ n,\; n+1,\; \cdots{} \} \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; x \not\in{} \{ n,\; n+1,\; \cdots{} \} \end{cases} $
<img src="../../png/03.3. Discrete Distributions (4).png" width="40%" height="40%"></img>

#### **(WHY?)**  
#### 불공평한 동전을 던져 $ \; n $ 개의 앞면이 나올 때까지 시행할 경우, 확률 변수 $ \; X $를 시행한 횟수라고 한다면, 
#### (단, $ \; P(H) = p $)
#### $ S = \{ (H_{1}H_{2}\cdots{}H_{n-1}H_{n}),\; (T_{1}H_{1}H_{2}\cdots{}H_{n-1}H_{n}),\; (H_{1}T_{1}H_{2}\cdots{}H_{n-1}H_{n}),\;  \cdots{} \} $
#### $ X = \{ n,\; n+1,\; n+1,\; \cdots{} \} = \{ n,\; n+1,\; n+2,\; \cdots{} \} $
#### if $ \; X = n $ 
#### $ \hspace{0.3cm} (H_{1}H_{2}\cdots{}H_{n-1}H_{n}) = (H_{1}H_{2}\cdots{}H_{n-1})(H_{n}) = (㉠) \times{} (㉡) $
#### $ \hspace{0.3cm} ㉠ \; $ : $ \; n-1 $ 개 시행 중 앞면이 $ \; n-1 $ 개 나올 경우의 수 $ = C^{n-1}_{n-1} $ 
#### $ \hspace{0.3cm} ㉡ \; $ : $ \; 1 $ 개 시행 중 앞면이 $ \; 1 $ 개 나올 경우의 수 $ = 1 $ 
#### if $ \; X = n+1 $ 
#### $ \hspace{0.3cm} \begin{rcases} (T_{1}H_{1}H_{2}\cdots{}H_{n-1}H_{n}) = (T_{1}H_{1}H_{2}\cdots{}H_{n-1})(H_{n}) \;\; \\ (H_{1}T_{1}H_{2}\cdots{}H_{n-1}H_{n}) = (H_{1}T_{1}H_{2}\cdots{}H_{n-1})(H_{n}) \;\; \\ \cdots{} \;\; \\ (H_{1}H_{2}\cdots{}H_{n-1}T_{1}H_{n}) = (H_{1}H_{2}\cdots{}H_{n-1}T_{1})(H_{n}) \;\; \end{rcases} = (㉠) \times{} (㉡) $
#### $ \hspace{0.3cm} ㉠ \; $ : $ \; n $ 개 시행 중 앞면이 $ \; n-1 $ 개 나올 경우의 수 $ = C^{n-1}_{n} $ 
#### $ \hspace{0.3cm} ㉡ \; $ : $ \; 1 $ 개 시행 중 앞면이 $ \; 1 $ 개 나올 경우의 수 $ = 1 $ 
#### $ \cdots{} $
#### if $ \; X = x $
#### $ P(㉠) = C^{x-1}_{n-1}(1-p)^{x-n}(p)^{n-1} \;\; $ ($ \because{} \; $ Binomial Formula)
#### $ P(㉡) = 1 \times{} p $  
#### $ \therefore{} \; P(X=x) = P(㉠)P(㉡) = C^{x-1}_{n-1}(1-p)^{x-n}(p)^{n-1} \times{} (p) = C^{x-1}_{n-1}(1-p)^{x-n}(p)^{n} $

#### **(PLUS)** 이 때 확률 변수 $ \; X $ 는 **특정 결과가 $ \; n $ 회 나올 때까지 시행하는 횟수**로 정의되며,

> ## **초기하 분포(Hypergeometric Distribution)**

#### **(1) 정의** : 
#### 전체 크기가 $ \; N $, 특정 요소의 크기가 $ \; k $ 인 서로 다른 두 요소를 갖는 집합에서 $ \; n $개의 표본을 추출했을 때
#### 특정 요소를 **비복원추출**하는 확률 변수가 따르는 분포
#### = **[CONTENS]**
#### $ \Rightarrow{} P(X=x) = \begin{cases} {{C^{k}_{x}C^{n-k}_{n-x}}\over{C^{N}_{n}}}, \;\; x \in{} \{ x \in{} \mathbb{Z} \; | \; \text{max}(0,\; n-N+k) \leq{} x \leq{} \text{min}(n,\; k) \} \\ 0, \;\;\;\;\;\;\;\;\;\; x \not\in{} \{ x \in{} \mathbb{Z} \; | \; \text{max}(0,\; n-N+k) \leq{} x \leq{} \text{min}(n,\; k) \} \end{cases} $

#### **(WHY?)**
#### **[CONTENTS]**

#### **(PLUS)** 이 때 확률 변수 $ \; X $ 는 전체 $ \; N $ 개 중 $ \; n $ 개의 표본 중 **특정 그룹을 뽑는 경우의 수**로 정의되며, 
#### 전체 집단의 개수 $ \; N $ 과 특정 그룹의 개수 $ \; k $, 표본 추출한 개수 $ \; n $ 을 모수로 하는 초기하 분포를 따름
#### $ \Rightarrow{} X \sim{} \text{Hypergeometric}(N, k,\; n) $

> ## **포아송 분포(Poisson Distribution)**

#### **(1) 정의** : 특정 구간(시간/공간)동안 **독립적이고 일정하게** 발생한 사건의 빈도(횟수)인 확률변수가 따르는 분포
#### = **[CONTENTS]**
#### $ \Rightarrow{} P(X=x) = \begin{cases} {{e^{-\lambda{}}\lambda{}^{x}}\over{x!}}, \;\; x \in{} \{ 0,\; 1,\; \cdots{} \} \\ 0, \;\;\;\;\;\;\;\; x \not\in{} \{ 0,\; 1,\; \cdots{} \} \end{cases} $
#### ※ $ \lambda{} $ : 사건 발생의 **평균** 빈도 $ \times{} $ 특정구간 $ = $ 특정 구간에서 발생한 이벤트의 평균 빈도

#### $ e = \displaystyle\lim_{n \rightarrow{} \infty{}} (1 + \frac{1}{n})^{n} = \displaystyle\sum^{\infty{}}_{n=1} \frac{1}{n!} \approx{} 2.71828 $

#### **(PLUS)** 
#### 이 때 확률 변수 $ \; X $ 는 **특정 구간에서 발생한 사건의 개수**로 정의되며,
#### 특정 구간에서 발생한 이벤트의 평균 빈도 $ \; \lambda{} $ 를 모수로 하는 포아송 분포를 따름

#### $ E(X) = \displaystyle\sum^{\infty{}}_{x=0} xP(X=x) = \displaystyle\sum^{\infty{}}_{x=0} x {{e^{-\lambda{}}\lambda{}^{x}}\over{x!}} $

#### $ \hspace{0.75cm} = e^{-\lambda{}} \lambda{}^{2} e^{\lambda{}} - \lambda{} = \lambda{}^{2} + \lambda{} $

#### **(PLUS) 포아송 분포와 이항 분포의 관계** :
