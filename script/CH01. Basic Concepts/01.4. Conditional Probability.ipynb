# CH01.4. **Conditional Probability**

> ## **조건부확률(Conditional Probability)**

#### **(1) 정의** : 사건 $ \; B $ 가 **이미 발생했을 때** 사건 $ \; A $ 의 확률 
#### = the conditional probability of $ \; A \; $ given that $ \; B \; $ has occured
#### $ \Rightarrow{} \; P(A|B) = {{P(A\cap{}B)}\over{P(B)}} \;\; $ ($ \; P(B) > 0 $ ) 

#### **(PLUS)** 조건부확률은 표본 공간이 사건 $ \; B $ 로 축약됐다고 이해할 수 있음

#### **(2) 예시** :
#### &nbsp;&nbsp; ① 성심당 빵을 먹을 확률 : $ \; P(B) $, 
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 대전에 갈 확률 : $ \; P(D) $, 
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 대전에 갔을 때 성심당 빵을 먹을 확률 : $ \; P(B|D) $
#### &nbsp;&nbsp; ② 주사위를 던졌을 때 소수(odd number)를 뽑을 확률 : $ \; P(A) = {{\{ 1,\; 3,\; 5 \}}\over{\{1,\; 2,\; 3,\; 4,\; 5,\; 6\}}} = {{3}\over{6}} = {{1}\over{2}} $, 
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 주사위를 던졌을 때 3 이하를 뽑을 확률 : $ \; P(B) = {{\{ 1,\; 2,\; 3 \}}\over{\{1,\; 2,\; 3,\; 4,\; 5,\; 6\}}} = {{3}\over{6}} = {{1}\over{2}} $,
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 주사위를 던졌을 때 3 이하를 뽑았고 소수를 뽑을 확률 : $ \; P(A|B) = {{P(A\cap{}B)}\over{P(B)}} = {{|A\cap{}B|}\over{|B|}} = {{\{ 1 \}}\over{\{ 1,\; 2,\; 3 \}}} = {{1}\over{3}} $ 

<b></b>

> ## (Theorem) **조건부확률의 연쇄법칙(Chain rule)**

#### $ n $ 개 사건의 조건부확률은 아래와 같다.
#### $ P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1})P(A_{2}|A_{1})P(A_{3}|A_{1}\cap{}A_{2})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1}) $
#### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = P(A_{1})P(A_{3}|A_{1})P(A_{2}|A_{1}\cap{}A_{3})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1}) $
#### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; \cdots{} $
#### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = P(A_{2})P(A_{1}|A_{2})P(A_{3}|A_{1}\cap{}A_{2})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1}) $
#### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; \cdots{} $
#### $ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; = P(A_{n})P(A_{n-1}|A_{n})P(A_{n-2}|A_{n-1}\cap{}A_{n})\cdots{}P(A_1|A_{2}\cap{}A_{3}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n}) $

#### **(WHY?)** 
#### if $ \; B_{1} = ( A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n} ) $
#### $ P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1}\cap{}B_{1}) = P(A_{1}|B_{1})P(B_{1}) $
#### if $ \; B_{2} = ( A_{3}\cap{}A_{4}\cap{}\cdots{}\cap{}A_{n} ), \; B_{1} = A_{2}\cap{}B_{2} $
#### $ P(B_{1}) = P(A_{2}\cap{}B_{2}) = P(A_{2}|B_{2})P(B_{2}) $
#### $ \cdots{} $ 
#### if $ \; B_{n-1} = A_{n}, \; B_{n-1} = A_{n-1}\cap{}B_{n-1} $
#### $ P(B_{n-1}) = P(A_{n}\cap{}B_{n}) = P(A_{n}|B_{n})P(B_{n}) $
#### $ P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1}|B_{1})P(A_{2}|B_{2})\cdots{}P(A_{n}|B_{n-1}) $
#### $ \therefore{} \; P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1})P(A_{2}|A_{1})P(A_{3}|A_{1}\cap{}A_{2})\cdots{}P(A_{n}|A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}\cdots{}\cap{}A_{n-1})  $

#### **(PLUS)** 해당 등식의 개수는 $ \; n! = n\times{}(n-1)\times{}\cdots{}\times{}1 $ 개임

<b></b>

> ## **독립(Independence)**

#### **(1) 정의** : 한 사건이 일어날 확률이 다른 사건이 일어날 확률에 영향을 미치지 않는 관계
#### = relationship where the occurrence of one event does not affect the chances of the occurrence of the other event
#### $ \Rightarrow{} \; P(A\cap{}B) = P(A)P(B) $

#### **(PLUS)** 두 사건이 독립일 경우, 각 사건의 조건부확률은 아래와 같음 
#### $ P(A|B) = {{P(A\cap{}B)}\over{P(B)}} = P(A) \Leftrightarrow{} P(B|A) = {{P(A\cap{}B)}\over{P(A)}} = P(B) \Leftrightarrow{} P(A\cap{}B) = P(A)P(B) $

#### **(2) 예시** :
#### &nbsp;&nbsp; 주사위 $ \; 1 $ 회 던져 짝수를 갖는 사건 $ \; A $ 와 $ \; 3 $ 이하를 갖는 사건 $ \; B $ 이 있을 때  
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ S = \{ 1,\; 2,\; 3,\; 4,\; 5,\; 6 \},\; A = \{ 2,\; 4 \},\; B = \{ 1,\; 2,\; 3 \},\; A\cap{}B = \{ 2 \} $
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ P(A\cap{}B) = {{1}\over{6}} = P(A)P(B) = {{2}\over{6}}\times{}{{3}\over{6}} = {{1}\over{6}} $
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ \therefore{} $ 두 이벤트 $\;  A,\; B $ 는 독립 관계이다.

#### **(PLUS)** 사건의 독립은 벤다이어그램으로는 파악하기 힘듬

#### **(3) 독립에 따른 집합 관계** : 두 사건 $ \; A,\; B $ 가 독립 관계일 때 

#### &nbsp;&nbsp; ① 사건 $ \; A $ 와 $ B^{c} $ 은 독립 관계임
#### &nbsp;&nbsp; ② 사건 $ \; A^{c} $ 와 $ B $ 은 독립 관계임
#### &nbsp;&nbsp; ③ 사건 $ \; A^{c} $ 와 $ B^{c} $ 은 독립 관계임

#### **(4) $ \; n $ 개 이상의 사건에 대한 독립 조건** :
#### $ P(A_{1}\cap{}A_{2}\cap{}A_{3}) = P(A_{1})P(A_{2}|A_{1})P(A_{3}|A_{1}\cap{}A_{2}) = P(A_{1})P(A_{2})P(A_{3}) $
#### $ P(A_{1}\cap{}A_{2}\cap{}A_{3}\cap{}A_{4}) = P(A_{1})P(A_{2})P(A_{3})P(A_{4}) $
#### $ \cdots{} $
#### $ \therefore{} \; P(A_{1}\cap{}A_{2}\cap{}\cdots{}\cap{}A_{n}) = P(A_{1})P(A_{2})\cdots{}P(A_{n}) $

#### **(PLUS)** $ \; n $ 개 사건이 독립 관계이려면, **모든 두 사건의 쌍들이 독립** 관계여야함

#### **(PLUS) $ \; n $ 개의 사건이 독립관계일 때 확률의 합집합** :

#### $ P(A_{1}\cup{}A_{2}\cup{}\cdots{}\cup{}A_{n}) = P((A_{1}^{c}\cap{}A_{2}^{c}\cap{}\cdots{}\cup{}A_{n}^{c})^{c}) $ ($ \because{} $ 드모르간의 법칙)
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ = 1 - P(A_{1}^{c}\cap{}A_{2}^{c}\cap{}\cdots{}\cup{}A_{n}^{c}) $ 

<b></b>

> ## **(Theorem)** 독립과 배반(서로소 집합)의 관계

#### 두 사건 $ A,\; B $ 의 확률 $ \; P(A) \neq{} 0,\; P(B) \neq{} 0 $ 이고 두 사건이 배반 관계라면, 사건 $ A,\; B $ 는 **독립 관계가 아니다**.

#### **(PLUS) 독립 관계(사건)와 배반 관계의 의미**
#### 만약 두 사건 $ \; A,\; B $ 가 독립 관계인 것은 사건 $ \; A $ 는 사건 $ B $ 에 대한 어떠한 정보를 전달하지 않는다는 의미이고
#### 만약 두 사건 $ \; A,\; B $ 가 배반 관계(disjoint set)인 것은 사건 $ \; A $ 와 사건 $ B $ 가 동시에 일어날 수 없다는 의미이다.
#### (물리적으로 독립상황인 것과는 다른 의미임)

#### **(WHY?)** 주사위를 $ \; 1 $ 회 던져 $ \; 3 \; $ 이하의 눈을 갖는 사건 $ \; A $ 와 $ \; 3 \; $ 초과의 눈을 갖는 사건 $ \; B $ 가 있을 때
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ S = \{ 1,\; 2,\; 3,\; 4,\; 5,\; 6 \},\; A = \{ 1,\; 2,\; 3 \},\; B = \{ 4,\; 5,\; 6 \},\; A\cap{}B = \empty{} $
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ P(A) = {{1}\over{3}},\; P(B) = {{1}\over{3}}, \;P(A\cap{}B) = 0,\; P(A|B) = P(B|A) = 0 $
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ P(A)P(B) \neq{} P(A|B) \; $ or $ \; P(A)P(B) \neq{} P(B|A) $
#### &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $ \therefore{} $ 두 이벤트 $\;  A,\; B $ 는 독립 관계가 아니다.

<b></b>

> ## **(Theorem)** 전체 확률의 법칙(Law of Total Probability)

#### 만약 사건 $ \; B_{1},\; B_{2},\; \cdots{} $ 가 표본 공간의 분할(partition)이고 사건 $ \; B_{i} $ 의 합집합이 표본공간일 때
#### 즉 $ \; B_{i}\cap{}B_{j} = \empty{}, \; \cup{}_{i=1}B_{i} = S $ 일 때 사건 $ \; A $ 의 확률은 아래와 같음
#### $ P(A) = \sum{}_{i=1}P(A\cap{}B_{i}) = \sum{}_{i=1} P(A|B_{i})P(B_{i}) $

> ## **베이즈 정리(Bayes Theorem)**

#### **(1) 정의** : 사건 $ A,\; B $ 가 존재할 때 사건 $ A $ 의 조건부확률을 사건 $ B $ 의 조건부확률로 정의
#### $ \; → \; P(A|B) = {{P(A\cap{}B)}\over{P(B)}} = {{P(B|A)P(A)}\over{P(B)}} $

#### **(PLUS)** 일반적으로 베이즈 정리는 시간 종속적인 사건의 전후 확률에 대하여 구하고자할 때 활용

#### **(2) 정리의 확장** : 
#### 사건 $ \; B_{1},\; B_{2},\; \cdots{} $ 가 서로 분활 관계이고, 개별 사건 $ \; B_{i} $ 의 합집합이 표본공간일 때 
#### $ P(B_{i}|A) = {{P(A|B_{i})P(B_{i})}\over{P(A)}} = {{P(A|B_{i})P(B_{i})}\over{\sum{}_{k=1}P(A|B_{k})P(B_{k})}} $

#### **(3) 교집합과 조건부확률의 해석** :
#### (한달 중) 비가 오는 날을 사건 $ \; A $, 우산을 챙긴 날을 사건 $ \; B $ 라고 할 때
#### $ P(B|A) \; $ : 한달 중 비가 올 때 우산을 챙길 확률, $ \; P(A) \; $ : 한달 중 비가 오는 날의 확률
#### $ P(B|A)P(A) \; $ : 비가 오는 날 우산을 들고 가는 날의 확률 $ \; = P(A\cap{}B) $
#### $ \therefore{} P(A\cap{}B) = P(A|B)P(B) = P(B|A)P(A) $

> ## **조건부 독립(Conditional Independence)**

#### **(1) 정의** : 사건 $ A,\; B $ 가 특정 사건 $ \; C $ 가 발생했을 때 독립일 경우의 관계
#### = Two events $ \; A,\; B \; $ are conditionally independent given an event $ \; C $ with $ \; P(C) \neq{} 0 $
#### $ \; → \; P(A\cap{}B|C) = P(A|C)P(B|C) $
