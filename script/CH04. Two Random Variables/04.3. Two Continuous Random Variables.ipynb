#### **(3) 성질** :
#### $ \hspace{0.15cm} $ ① $ 0 \leq{} P(X=x, Y=y) \leq{} 1 \;\; $ for all $ \; x,\; y $
#### $ \hspace{0.15cm} $ ② $ \displaystyle\int^{\infty{}}_{-\infty{}} \displaystyle\int^{\infty{}}_{-\infty{}} f(X=x,\; Y=y) \text{dx}\text{dy} = \displaystyle\int{} \displaystyle\int_{(X,\;Y)(S)} f(X=x,\; Y=y) \text{dx}\text{dy} = 1 $

#### **(4) 주변 확률밀도함수(Marginal PDF)** : **하나의 확률변수에 대해** 다른 확률변수 관계없이 정의하는 함수
#### $ \hspace{0.15cm} $ ① 확률변수 $ \, X $ 의 확률밀도함수
#### $ \hspace{0.6cm} f(X=x) = \displaystyle\int^{\infty{}}_{-\infty{}} f(X=x,\, Y=y) \text{dy} = \displaystyle\int_{y\in{}Y(S)} f(X=x,\, Y=y) \text{dy} $
#### $ \hspace{0.15cm} $ ② 확률변수 $ \, Y $ 의 확률밀도함수
#### $ \hspace{0.6cm} f(Y=y) = \displaystyle\int^{\infty{}}_{-\infty{}} f(X=x,\, Y=y) \text{dx} = \displaystyle\int_{x\in{}X(S)} f(X=x,\, Y=y) \text{dy} $

> ## **결합 누적밀도함수(Joint Cumulative Density Function; Joint CDF)**

#### **(1) 정의** : **[CONTENTS]**
#### $ \Rightarrow{} F(X=x,\; Y=y) = P(X\leq{}x,\; Y\leq{}y) = \displaystyle\int^{y}_{-\infty{}} \displaystyle\int^{x}_{-\infty{}} f(X=x,\; Y=y) \text{dx}\text{dy} $ 

#### **(2) 성질** :
#### $ \hspace{0.15cm} $ ① $ F(X\leq{}\infty{},\; Y\leq{}\infty{}) = 1 $
#### $ \hspace{0.15cm} $ ② $ F(X\leq{}-\infty{},\; Y\leq{}y) = 0 $
#### $ \hspace{0.15cm} $ ③ $ F(X\leq{}x,\; Y\leq{}-\infty{}) = 0 $
#### $ \hspace{0.15cm} $ ④ $ F(X\leq{}-\infty{},\; Y\leq{}-\infty{}) = 0 $

#### **(3) 주변 누적분포함수(Marginal CDF)** : **하나의 확률변수에 대해** 다른 확률변수 관계없이 정의하는 함수
#### $ \hspace{0.15cm} $ ① 확률변수 $ \, X $ 의 누적분포함수
#### $ \hspace{0.6cm} F(X=x) = P(X\leq{}x,\, Y\leq{}\infty{}) = \displaystyle\sum_{y_{k}\in{}Y(S)} P(X\leq{}x,\, Y=y_{k}) = P(X\leq{}x) $
#### $ \hspace{0.15cm} $ ② 확률변수 $ \, Y $ 의 누적분포함수
#### $ \hspace{0.6cm} F(Y=y) = P(X\leq{}\infty{},\, Y\leq{}y) = \displaystyle\sum_{x_{i}\in{}X(S)} P(X=x_{i},\, Y\leq{}y) = P(Y\leq{}y) $

> ## **조건부 확률밀도함수(Conditional Probability Density Function)**

#### **(1) 정의** : 결합 확률밀도함수에서 **하나의 확률변수가 특정 값을 취했을 때** 다른 확률변수에 대한 함수
#### $ \Rightarrow{} f_{X|Y}(x|y_{k}) = f(X=x|Y=y_{k}) = \frac{f(X=x,\; Y=y_{k})}{f(Y=y_{k})} = \frac{f(\{X=x\} \cap{} \{Y=y_{k}\})}{f(Y=y_{k})} \;\; \text{ for any } \; x \in{} X(S), \;\; y_{k} \in{} Y(S) $

#### **(`WHY?`)**
#### $ \hspace{0.15cm} P(a\leq{}X\leq{}b|Y=c) = \displaystyle\lim_{\epsilon{}\rightarrow{}0} P(a\leq{}X\leq{}b|c-\epsilon{}\leq{}Y\leq{}c+\epsilon{}) $
#### $ \hspace{3.4cm} = \displaystyle\lim_{\epsilon{}\rightarrow{}0} \frac{P(a\leq{}X\leq{}b,\; c-\epsilon{}\leq{}Y\leq{}c+\epsilon{})}{P(c-\epsilon{}\leq{}Y\leq{}c+\epsilon{})} $
#### $ \hspace{3.4cm} = \displaystyle\lim_{\epsilon{}\rightarrow{}0} \frac{\displaystyle\int^{b}_{a}\displaystyle\int^{c+\epsilon{}}_{c-\epsilon{}}f(X=x,\; Y=y) \text{dy}\text{dx}}{\displaystyle\int^{\infty{}}_{-\infty{}}\displaystyle\int^{c+\epsilon{}}_{c-\epsilon{}}f(X=x,\; Y=y) \text{dy}\text{dx}} $
#### $ \hspace{3.4cm} \approx{} \displaystyle\lim_{\epsilon{}\rightarrow{}0} \frac{\displaystyle\int^{b}_{a}\displaystyle\int^{c+\epsilon{}}_{c-\epsilon{}}f(X=x,\; Y=c) \text{dy}\text{dx}}{\displaystyle\int^{\infty{}}_{-\infty{}}\displaystyle\int^{c+\epsilon{}}_{c-\epsilon{}}f(X=x,\; Y=c) \text{dy}\text{dx}} $
#### $ \hspace{3.4cm} = \displaystyle\lim_{\epsilon{}\rightarrow{}0} \frac{\displaystyle\int^{b}_{a}2\epsilon{}f(X=x,\; Y=c) \text{dx}}{\displaystyle\int^{\infty{}}_{-\infty{}}2\epsilon{}f(X=x,\; Y=c) \text{dx}} \;\; $ ($ \because{} \, c \, \text{ is constant}. $)
#### $ \hspace{3.4cm} = \frac{\displaystyle\int^{b}_{a}f(X=x,\; Y=c) \text{dx}}{\displaystyle\int^{\infty{}}_{-\infty{}}f(X=x,\; Y=c) \text{dx}} $
#### $ \hspace{3.4cm} = \displaystyle\int^{b}_{a} \frac{f(X=x,\; Y=c)}{f(Y=c)} \text{dx} $

#### **(2) 예시** :
#### 연속확률변수 $ \, X,\, Y $ 에 대하여 아래와 같이 결합 확률밀도함수가 정의될 때
#### $ f(X=x,\;Y=y) = \begin{cases} \frac{6}{5}(x+y^{2}), \;\; 0 \leq{}x\leq{}1, \;\; 0\leq{}y\leq{}1 \\ 0, \;\;\;\;\;\;\;\;\;\;\;\;\;\;\; \text{otherwise.} \end{cases} $
#### $ \hspace{0.15cm} $ ① 조건부 확률밀도함수 $ \, f(Y=y|X=0.5) $ 는?
#### $ \hspace{0.45cm} \cdot{} \, f(X=x) = \displaystyle\int^{1}_{0}f(X=x,\, Y=y)\text{dy} = \displaystyle\int^{1}_{0}\frac{6}{5}(x+y^{2})\text{dy} $
#### $ \hspace{2.175cm} = \frac{6}{5}(\displaystyle\int^{1}_{0}x\text{dy} + \displaystyle\int^{1}_{0}y^{2}\text{dy}) = \frac{6}{5}( x(1-0) + \begin{bmatrix} \frac{y^{3}}{3}\end{bmatrix}^{1}_{0}) = \frac{6}{5}(x+\frac{1}{3}) $
#### $ \hspace{0.45cm} \therefore{} f(X=0.5,\; Y=y) = \frac{\frac{6}{5}(\frac{1}{2}+y^{2})}{\frac{6}{5}(\frac{1}{2}+\frac{1}{3})} = \frac{\frac{1}{2}+y^{2}}{\frac{1}{2}+\frac{1}{3}} = \frac{3}{5} + \frac{6}{5}y^{2} $
#### $ \hspace{0.15cm} $ ② 확률 $ \, P(Y\leq{}0.5|X=0.5) $ 의 값은?
#### $ \hspace{0.45cm} \cdot{} \, P(Y\leq{}0.5|X=0.5) = \displaystyle\int^{0.5}_{0} (\frac{3}{5} + \frac{6}{5}y^{2})\text{dy} $
#### $ \hspace{0.45cm} $ **[LATEX]**

> ## **조건부 기대값(Conditional Expectation)**

#### **(1) 정의** : 하나의 연속확률변수가 특정 값을 취할 때 다른 연속확률변수의 기대값
#### $ \Rightarrow{} E(X=x|Y=y_{k}) = \displaystyle\int_{x_{i}\in{}X(S)} x_{i} f(X=x|Y=y_k) \text{dx} $ 

<b></b>

> ## **조건부 분산(Conditional Variance)**

#### **(1) 정의** : 조건부 기대값으로부터 얼마나 떨어져 있는지를 측정한 값(지표)
#### $ \Rightarrow{} \text{var}(X=x|Y=y_{k}) = E[(X-E(X))^{2}|Y=y_{k}] $ 
#### $ \hspace{3.5cm} = E(X^{2}|Y=y_{k}) - E(X)^{2}  $

<b></b>

> ## **연속결합확률변수의 독립(Independnece of Joint Continuous RV)**

#### **(1) 정의** : 한 이산확률변수가 일어날 확률이 다른 이산확률변수가 일어날 확률에 영향을 미치지 않는 관계
#### $ \Rightarrow{} f(X=x,\; Y=y) = f(X=x)f(Y=y), \;\; \forall{} \; x \in{} X(S), \;\; y \in{} Y(S) $
#### $ \Leftrightarrow{} f(X=x|Y=y_{k}) = \frac{f(X=x,\; Y=y_{k})}{f(Y=y_{k})} = \frac{f(X=x)f(Y=y_{k})}{f(Y=y_{k})} = f(X=x), \;\; \forall{} \; x \in{} X(S), \;\; y \in{} Y(S) $
#### $ \Leftrightarrow{} f(Y=y|X=x_{i}) = \frac{f(X=x_{i},\; Y=y)}{f(X=x_{i})} = \frac{f(X=x_{i})f(Y=y)}{f(X=x_{i})} = f(Y=y), \;\; \forall{} \; x \in{} X(S), \;\; y \in{} Y(S) $

#### **(`PLUS`)** 결합 누적분포함수에서도 동일하게 적용됨
#### $ \hspace{0.15cm} \Rightarrow{} F(X=x,\; Y=y) = F(X=x)F(Y=y), \;\; \forall{} \; x \in{} X(S), \;\; y \in{} Y(S) $

<b></b>

> ## **연속결합확률변수의 합성함수(Composite Functions of Continous Joint RV)**

#### **(1) 정의** : 연속결합확률변수를 다른 실수 공간으로 대응시킨 **확률 변수**
#### $ \Rightarrow{} Z = g \circ{} (X,Y) = g(X, Y) : S \rightarrow{} \mathbb{R} \;\; $ (※ $ g : \mathbb{R}^2 \rightarrow{} \mathbb{R} $)
#### **[GRAPH]**

#### **(2) 합성함수의 확률** :
#### $ \Rightarrow{} P(Z=z_{m}) = f(g(X, Y)=z_{m}) = \displaystyle\int_{(x_{i},\;y_{k}) \in{} A_{z_{m}} } f_{XY}(x_{i},\; y_{k}) \text{dx} \text{dy} $
#### $ \hspace{0.15cm} \text{where } \, A_{z_{m}} = \{ (x_{i},\; y_{k}) \; | \; Z(X,\; Y) = z_{m} \;\; \& \;\; (x_{i},\; y_{k}) \in{} (X,\;Y)(S) \} $

#### **(3) 합성함수의 기댓값** : 
#### $ \Rightarrow{} E(Z) = \displaystyle\int_{z_{m}\in{}Z(X,\;Y)} z_{m} \displaystyle\int_{(x_{i},\; y_{k})\in{}A_{z_{k}}} P(X=x_{i},\; Y=y_{k}) \text{dx}\text{dy} $
#### $ \hspace{1.3cm} = \displaystyle\int_{z_{m}\in{}Z(X,\;Y)} \displaystyle\int_{(x_{i},\; y_{k})\in{}A_{z_{k}}} z_{m} P(X=x_{i},\; Y=y_{k}) \text{dx}\text{dy} $
#### $ \hspace{1.3cm} = \displaystyle\int_{(x_{i},\; y_{k})\in{}(X,\;Y)(S)} g(X=x_{i},\; Y=y_{k})P(X=x_{i},\; Y=y_{i}) \text{dx}\text{dy} $

#### **(4) 합성함수의 분산** :
#### $ \Rightarrow{} \text{var}(Z) = E(Z^{2}) - E(Z)^{2} = \displaystyle\int_{(x_{i},\;y_{k})\in{}(X,\;Y)(S)} g(x_{i},\; y_{k})^{2} f(X=X_{i},\; Y=y_{k}) \text{dx}\text{dy} - E(Z)^{2} $
